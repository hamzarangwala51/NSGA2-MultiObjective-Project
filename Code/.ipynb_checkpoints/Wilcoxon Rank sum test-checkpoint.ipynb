```python
from scipy.stats import ranksums
```

```python
def wilcoxon_rank_sum_test(values1, values2):
    """
    Performs the Wilcoxon Rank-Sum Test on two sets of values.
    
    Parameters:
        values1 (list or array-like): The first set of values (e.g., 10 values for the first method).
        values2 (list or array-like): The second set of values (e.g., 10 values for the second method).
    
    Returns:
        p_value (float): The p-value from the Wilcoxon Rank-Sum Test.
    """
    # Perform the Wilcoxon Rank-Sum Test
    stat, p_value = ranksums(values1, values2)
    
    # Return the p-value
    return p_value
```
